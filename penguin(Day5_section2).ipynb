```python

import numpy as np
numerical_cols=df.select_dtypes(include=np.number).columns
for col in numerical_cols:
  plt.figure(figsize=(8,6))
  sns.boxplot(y=df[col])
  plt.title(f'Box plot of {col}')
  plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df
```

```python
import pandas as pd
df=pd.get_dummies(data=df,columns=['island'],prefix="island",drop_first=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
df["body_mass_g"]=std.fit_transform(df[["body_mass_g"]])
df
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
df["flipper_length_mm"]=std.fit_transform(df[["flipper_length_mm"]])
df
```

```python
df.isnull().sum()
```

```python
bol_cols=df.select_dtypes(include='bool').columns
df[bol_cols]=df[bol_cols].astype(int)
```

```python
df
```

```python
y=df['species']
x=df.drop(['species'],axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_predict=lr_model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
confusion_matrix(y_test,y_predict)
```

```python
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
y_pred=dt_model.predict(x_test)
print(accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```

```python
dt_model=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10,random_state=42,min_samples_leaf=10)
dt_model.fit(x_train,y_train)
y_pred=dt_model.predict(x_test)
print(accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
y_predict=rf_model.predict(x_test)
print(accuracy_score(y_test,y_predict))
confusion_matrix(y_test,y_predict)
```

```python
rf_model=RandomForestClassifier(n_estimators=20,max_depth=10,
                                min_samples_split=3)
rf_model.fit(x_train,y_train)
y_predict=rf_model.predict(x_test)
print(accuracy_score(y_test,y_predict))
confusion_matrix(y_test,y_predict)
```
