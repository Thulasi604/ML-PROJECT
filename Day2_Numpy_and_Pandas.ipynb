```python
df=df.drop(columns=['deck'])
df.shape
```

```python
#Remove null values
df.dropna(subset=['embarked','embark_town'],inplace=True)
print(df.isnull().sum())

```

```python
age_mean=df['age'].mean()
df['age']=df['age'].fillna(age_mean)
```

```python
df['family_size']=df['sibsp']+df['parch']
```

```python
#creating and modifying columns
df['age_group']=np.where(df['age']<18,"child","adult")
```

```python
df['fare_per_person']=df['fare']/(df['family_size']+1)
```

```python
df
```

```python
#Rename
df.rename(columns={'pclass':'Passenger_Class'},inplace=True)
df
```

```python
#Filtering and conditional selection
```

```python
df[df['age']>df['age'].mean()]
```

```python
#female passengers in class
```

```python
df[(df['sex']=='female')&(df['Passenger_Class']==1)]
```

```python
#survivors only
df[df['survived']==1]
```

```python
#survival rate by gender
```

```python
df.groupby('sex')['survived'].mean()
```

```python
#Average fare by class
df.groupby('Passenger_Class')['fare'].mean()
```

```python
#highly correlated variables
```

```python
from numpy._core import numerictypes
df.corr(numeric_only=True)
```
